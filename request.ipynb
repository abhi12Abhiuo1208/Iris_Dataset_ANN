# to see how we can use a code for pulling request

```python
import requests
```

```python
flower_example = {"sepal_length": 5.1, 
                "sepal_width": 3.5, 
                "petal_length": 1.4, 
                "petal_width": 0.2}
```

```python
result = requests.post("http://localhost:5000/api/flower", json = flower_example)
```

```python
result.status_code
```

```python
print(result.json())
print(result.text)
```
